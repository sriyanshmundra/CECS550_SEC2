## Part 6

1. Matrix multiply `T` and `img_average_channel` and store the result in `transformed_image`. Check the shape of the result.
2. Plot `transformed_image` and describe the pattern you see.
3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image.

```python
#1. Matrix multiplication
transformed_image = T @ img_average_channel
print(transformed_image.shape)

#2. Plotting
plt.imshow(transformed_image)

'''
Since T is the transformation of the image, it rearranges the rows of the original img_average_channel. This results
in the random shuffling of rows and the image looks distorted.
'''

#3.
'''
In the matrix multiplication above a row is selected from T and then linearly combined with the rows of the
img_average_channel. The specific row order from T determines the pattern of rearrangement.
As the rows of T are shuffled randomly, we see a distorted or scrambled version of the original image.
'''
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
#1. T is invertible if the determinant is non zero. So we calculate it below

# Calculate the determinant of T
det_T = np.linalg.det(T)

if det_T != 0:
    # 2. Compute the inverse of T
    Tinv = np.linalg.inv(T)
    print("Matrix T is invertible.")
else:
    print("Matrix T is not invertible")

# We Perform the inverse operation
retrieved_image = Tinv @ transformed_image

plt.imshow(retrieved_image)

#We can observe that the inverted matrix is retrieved and is same as the original image
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image.

```python
og_image = img_average_channel.copy()
image_shape = og_image.shape

#created identity matrix
identity_matrix = np.eye(image_shape[0],image_shape[1])

#shuffling rows
row_indices = np.arange(image_shape[0])
np.random.shuffle(row_indices)
T_rows = identity_matrix[row_indices]

#shuffling columns
column_indices = np.arange(image_shape[1])
np.random.shuffle(column_indices)
T_columns = identity_matrix[column_indices]

scrambled_image = T_rows @ og_image @ T_columns

#Inverse the rows and columns to recover the original image
T_columns_inv = np.linalg.inv(T_columns)
T_rows_inv = np.linalg.inv(T_rows)

recovered_image = T_rows_inv @ scrambled_image @ T_columns_inv

plt.subplot(1, 2, 1)
plt.imshow(scrambled_image)
plt.subplot(1, 2, 2)
plt.imshow(recovered_image)

```
